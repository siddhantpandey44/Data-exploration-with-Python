# Data Exploration Using Python

```python
#import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
#plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

*Read the dataset into a dataframe*

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

*Data Cleaning*

```python
for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} {}'.format(col,missing*100))
```

```python
df.dtypes
```

```python
df.dropna(subset='budget',inplace=True)
```

```python
df.dropna(subset='gross',inplace=True)
```

```python
for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} {}'.format(col,missing*100))
```

```python
df.head()
```

```python
df['budget'] = df['budget'].astype('int64')
```

```python
df['gross'] = df['gross'].astype('int64')
```

```python
df.dtypes
```

```python
df.head()
```

Are there any outliers?

```python
df.boxplot(column=['gross'])
```

```python
df.boxplot(column=['budget'])
```

```python
df.drop_duplicates()
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

**Data Analysis and Correlation**

```python
plt.scatter('gross','budget',data=df)
plt.xlabel('Gross')
plt.ylabel('Budget')
plt.title('Gross vs Budget')

plt.legend

plt.show()
```

```python
sns.regplot('gross','budget',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
df.corr(method='pearson') # pearson, kendall, spearman
```

```python
df.corr(method='kendall')
```

```python
df.corr(method='spearman')
```

```python
Correlation = df.corr(method='pearson')
sns.heatmap(Correlation, annot=True)
plt.xlabel('Movies Feratures')
plt.ylabel('Movies Feratures')

plt.title('Correlation Matrix among numerical features')
```

```python
df.head()
```

```python
df_numerized = df
```

```python
for col in df_numerized.columns:
    if (df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
    
df_numerized
```

```python
df_numerized.dtypes
```

```python
for col in df_numerized.columns:
    if (df_numerized[col].dtype == 'category'):
        df_numerized[col] = df_numerized[col].cat.codes
    
df_numerized
```

```python
Correlation = df_numerized.corr(method='pearson')
sns.heatmap(Correlation, annot=True)
plt.xlabel('Movies Feratures')
plt.ylabel('Movies Feratures')

plt.title('Correlation Matrix among numerical features')
```

```python
corr_matrix = df_numerized.corr()
corr_pairs = corr_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
